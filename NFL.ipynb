```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Stroe CSV into DataFrame
NFL_file=".//Resources/NFL Draft.csv"
NFL_data_df = pd.read_csv(NFL_file)
NFL_data_df.head()
```

```python
new_NFL_data_df = NFL_data_df[['Player','College/Univ','Year']].copy()
new_NFL_data_df.head()
```

```python
new_NFL_data_df = new_NFL_data_df.rename(columns={'Player':'PLAYER','College/Univ': 'COLLEGE','Year':'YEAR'})
new_NFL_data_df.head()
```

```python
#Stroe CSV into DataFrame
NCAAF_file=".//Resources/NCAA Football Champions.csv"
NCAAF_data_df = pd.read_csv(NCAAF_file)
NCAAF_data_df.head()
```

```python
new_NCAAF_data_df = NCAAF_data_df[['YEAR', 'CHAMPION']].copy()
new_NCAAF_data_df.head()
```

```python
#Connect to local database
```

```python
rds_connection_string = "root:Eight1one@localhost:3306/sports_db"
engine = create_engine(f'mysql+pymysql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
 #Use pandas to load csv converted DataFrame into database
```

```python
new_NFL_data_df.to_sql(name='NFL_data', con=engine, if_exists='append', index=False)
```

```python
new_NCAAF_data_df.to_sql(name='NCAAF_data', con=engine, if_exists='append', index=False)
```
